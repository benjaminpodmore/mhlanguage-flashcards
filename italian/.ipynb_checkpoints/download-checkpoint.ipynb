```python
import requests
import csv
import pandas as pd
```

```python
chapters = requests.get('https://mhe-language-lab.azurewebsites.net/api/GetSubMenus?parentID=87414').json()

for idx, chapter in enumerate(chapters):
    title = chapter['MenuTitle']
    chapter_id = chapter['Menu_ID']
    print(title)

    chapter_sections = requests.get(f'https://mhe-language-lab.azurewebsites.net/api/GetSubMenus?parentID={chapter_id}').json()
    vocab = []
    for section in chapter_sections:
        chapter_section_id = section['Menu_ID']
        lists = requests.get(f'https://mhe-language-lab.azurewebsites.net/api/GetSubMenus?parentID={chapter_section_id}').json()
        flashcard_list = lists[0]
        flashcard_list_id = flashcard_list['Menu_ID']
        flashcards = requests.get(f'https://mhe-language-lab.azurewebsites.net/api/GetFlashCards?menuID={flashcard_list_id}').json()
        vocab.extend(flashcards)
        print(section['MenuTitle'])
    prefix = f"{idx:02}"
    
    cards = [[card['SideB'], card['SideA']] for card in vocab]
    pd.DataFrame(cards).to_csv(f"data/{prefix}. {title}.csv",index=False,header=False, sep="\t")
```
